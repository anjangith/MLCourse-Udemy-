```python
import pandas as pd
df = pd.read_excel('http://cdn.sundog-soft.com/Udemy/DataScience/cars.xls')
```

```python
df.head()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X = df[['Price','Cylinder','Liter','Doors']]
y = df['Mileage']
plt.show()

X[['Price','Cylinder','Liter','Doors']] = scale.fit_transform(X[['Price','Cylinder','Liter','Doors']].as_matrix())

print (X)

est = sm.OLS(y, X).fit()

est.summary()
```

```python
scaled = scale.transform([[17000,6,3.1,4]])
print(scaled)
predicted = est.predict(scaled[0])
print(predicted)
```
